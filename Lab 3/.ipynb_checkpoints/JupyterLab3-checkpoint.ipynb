```python
#original image
import cv2
img = cv2.imread('images/input.jpg')
cv2.imshow("image", img)
cv2.waitKey()
```

```python
#task 1 resize
import cv2
img = cv2.imread('images/input.jpg')
resized_img = cv2.resize(img, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_LINEAR)
cv2.imshow("resized", resized_img)
cv2.waitKey()
```

```python
#task 2 rotate
import cv2
img = cv2.imread('images/input.jpg')
rows, cols = img.shape[:2]
center = (cols / 2, rows / 2)
rot = cv2.getRotationMatrix2D(center, 120, 1.0)
rotated_img = cv2.warpAffine(img, rot, (cols, rows))
cv2.imshow("rotated", rotated_img)
cv2.waitKey()
```

```python
#task 3 shear
import cv2
img = cv2.imread('images/input.jpg')
shear = cv2.getRotationMatrix2D((0, 0), 0, 1.0)
shear[0, 1] = 0.2
shear[1, 0] = 0.2
rows, cols = img.shape[:2]
sheared_img = cv2.warpAffine(img, shear, (int(cols * 1.5), int(rows * 1.5)))
cv2.imshow("Sheared", sheared_img)
cv2.waitKey()
```

```python
#task 4 negative
import cv2
img = cv2.imread('images/input.jpg')
negative_img = cv2.bitwise_not(img)
cv2.imshow("Negative", negative_img)
cv2.waitKey()
```

```python
#task 5 log
import cv2
import numpy as np
img = cv2.imread('images/input.jpg')
log_img = np.uint8((255 / np.log(256)) * np.log1p(np.float32(img)))
cv2.imshow("Log", log_img)
cv2.waitKey()
```

```python
#task 6 power
import cv2
import numpy as np
img = cv2.imread('images/input.jpg')
power_img = np.uint8(255 * (img / 255.0) ** 2.0)
cv2.imshow("power", power_img)
cv2.waitKey()
```
